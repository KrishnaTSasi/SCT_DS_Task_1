```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import zipfile
```

```python
with zipfile.ZipFile("/content/API_SP.POP.TOTL_DS2_en_csv_v2_38144.zip",'r') as zip_ref:
  zip_ref.extract("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv","/content/")
```

```python
data=pd.read_csv("/content/API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv",skiprows=4)
data
```

```python
# Filter data for India
india_data=data[data['Country Name']=='India'].copy()
```

```python
# Filter data from 1960 to 2022
india_data=india_data.loc[:,'1960':'2022'].T
india_data.columns=['Population']
india_data.index.name='Year'
```

```python
# Rest index to make year a column
india_data.reset_index(inplace=True)
india_data['Year']=india_data['Year'].astype(int)
india_data['Population']=pd.to_numeric(india_data['Population'],errors='coerce')
```

BAR CHART

```python
# Plot the bar chart of Population Growth
plt.figure(figsize=(12, 6))
sns.barplot(data=india_data, x='Year', y='Population', color='cornflowerblue')
plt.xticks(rotation=45)
plt.title("India's Population Growth (1960–2022)", fontsize=15)
plt.xlabel("Year")
plt.ylabel("Population")
plt.tight_layout()
plt.show()
```

HISTOGRAM

```python
# Plot the histogram with KDE for population distribution
plt.figure(figsize=(10,8))
sns.histplot(data=india_data,x='Population',bins=15,kde=True,color='orchid')
plt.title("Histogram of India's Population (1960=2022)",fontsize=15)
plt.xlabel("Population")
plt.ylabel("Number of years")
plt.grid(True,linestyle='--',alpha=0.5)
plt.tight_layout()
plt.show()
```
